```python
def draw_health_bar(surface, x, y, health,max_health):
    # Set the size of the health bar
    BAR_LENGTH = 800
    BAR_HEIGHT = 20
    
    # Calculate the current health percentage
    health_percentage = health / max_health
    
    # Calculate the width of the health bar based on the current health percentage
    bar_width = int(BAR_LENGTH * health_percentage)
    
    # Set the color of the health bar based on the current health percentage
    if health_percentage > 0.6:
        color = GREEN
    elif health_percentage > 0.3:
        color = YELLOW
    else:
        color = RED
    
    # Draw the health bar
    pygame.draw.rect(surface, BLACK, [x, y, BAR_LENGTH, BAR_HEIGHT], 2)
    pygame.draw.rect(surface, color, [x+2, y+2, bar_width, BAR_HEIGHT-4])
    
```
